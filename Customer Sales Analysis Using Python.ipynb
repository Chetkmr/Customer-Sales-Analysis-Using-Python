```python
# import pandas

# data = pd.read_csv('sales data.csv', index_col=0, encoding='windows-1252')

import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('Diwali Sales Data.csv', index_col=0, encoding='windows-1252')
```

```python
data = data.dropna(axis=1)
```

```python
data.isnull().sum()
```

```python
data = data.dropna(axis=0)
```

```python
data.head(20)
```

```python
data.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.countplot(x='Zone', data=data)
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='Product_Category', data=data)
plt.xticks(rotation=70)
plt.show()
```

```python
state_sales = data.groupby('State')['Orders'].sum().sort_values(ascending=False)
plt.figure(figsize=(12,6))
state_sales.plot(kind='bar')
plt.ylabel("Total Sales Amount")
plt.title("Sales by State")
plt.show()
```

```python
sns.countplot(x='Gender', data=data)
plt.title("Orders by Gender")
plt.show()
```

```python
age_sales = data.groupby('Age Group')['Orders'].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
age_sales.plot(kind='bar', color='skyblue')
plt.ylabel("Total Amount")
plt.title("Sales by Age Group")
plt.show()
```
